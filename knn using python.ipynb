```python
#import the packages
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix


```

```python
#import the data
data = pd.read_csv("PROJECT - Feuille 1.csv")
```

```python
#select the columns that we will use
df = data[["City/locality", "Energy Consumption BEFORE intervention (actual)\nMWh/a","Energy Consumption AFTER intervention (actual)\nMWh/a","Predicted Energy Consumption AFTER intervention\nMWh/a","Value of EE investment\n€","Energy Performance Certificate rating AFTER intervention"]]
df
```

```python
#remove messing values
df = df.dropna()
df
```

```python
#build another column called Energy performance* that conatins the 3 categories 
df['Energy performance*']=df['Energy Performance Certificate rating AFTER intervention']
l=[]
for i in range(len(df['City/locality'])):
    p=((df['Energy Consumption BEFORE intervention (actual)\nMWh/a'].iloc[i]-float(df['Energy Consumption AFTER intervention (actual)\nMWh/a'].iloc[i].replace(',','').replace(' ','')))/df['Energy Consumption BEFORE intervention (actual)\nMWh/a'].iloc[i])*100
    l.append(p)
    if l[i]<60:
        df['Energy performance*'].iloc[i]='A'
    if l[i]<90 and l[i]>60:
        df['Energy performance*'].iloc[i]='AA'
    if l[i]>90:
        df['Energy performance*'].iloc[i]='AAA'
df = df.drop(['Energy Performance Certificate rating AFTER intervention','City/locality'], axis=1) 
df
```

```python
#convert the str values into float values 
for i in range(len(df['Energy Consumption BEFORE intervention (actual)\nMWh/a'])):
    df['Energy Consumption AFTER intervention (actual)\nMWh/a'].iloc[i]=float(df['Energy Consumption AFTER intervention (actual)\nMWh/a'].iloc[i].replace(',','').replace(' ',''))
    df['Predicted Energy Consumption AFTER intervention\nMWh/a'].iloc[i]=float(df['Predicted Energy Consumption AFTER intervention\nMWh/a'].iloc[i].replace(',','').replace(' ',''))
    df['Value of EE investment\n€'].iloc[i]=float(df['Value of EE investment\n€'].iloc[i].replace(',','').replace(' ',''))
df   
```

```python
#select the 4 first columns as features and the last column as lavel
X = df.iloc[:, :-1].values
y = df.iloc[:, 4].values
```

```python
#split the data into train and test using 80% for train 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#standarise the data
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#implement knn
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
#predict
y_pred = classifier.predict(X_test)
```

```python
#calculate the accuracy
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
